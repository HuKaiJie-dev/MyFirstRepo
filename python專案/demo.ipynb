## 這一個`sin圖表`的範例

```python
# prompt: create a sin plot

import matplotlib.pyplot as plt
import numpy as np

# Generate x values from 0 to 2π
x = np.linspace(0, 2 * np.pi, 100)

# Calculate corresponding y values for sin(x)
y = np.sin(x)

# Create the plot
plt.plot(x, y)

# Add labels and title
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.title("Sine Wave")

# Display the plot
plt.show()
```
